```python

def BE(Z,N,W,df):
    try:
        return float(df[(df['Z']==Z) & (df['N']==N) & (df['Wigner']==W)]['BE'])
    except:
        return None

def OneNSE(Z,N,W,df):
    try:
        return BE(Z,N,W,df) - BE(Z,N-1,W,df)
    except:
        return None
    
def OnePSE(Z,N,W,df):
    try:
        return BE(Z,N,W,df) - BE(Z-1,N,W,df)
    except:
        return None

def TwoNSE(Z,N,W,df):
    try:
        return BE(Z,N,W,df) - BE(Z,N-2,W,df)
    except:
        return None

def TwoPSE(Z,N,W,df):
    try:
        return BE(Z,N,W,df) - BE(Z-2,N,W,df)
    except:
        return None

def AlphaSE(Z,N,W,df):
    try:
        return BE(Z,N,W,df) - BE(Z-2,N-2,W,df) - 28.3
    except:
        return None

def TwoPSGap(Z,N,W,df):
    try:
        return TwoPSE(Z,N,W,df) - TwoPSE(Z+2,N,W,df)
    except:
        return None

def TwoNSGap(Z,N,W,df):
    try:
        return TwoNSE(Z,N,W,df) - TwoNSE(Z,N+2,W,df)
    except:
        return None

def DoubleMDiff(Z,N,W,df):
    try:
        return (TwoPSE(Z,N,W,df) - TwoPSE(Z,N-2,W,df))/4
    except:
        return None

def N3PointOED(Z,N,W,df):
    try:
        return (-1)**N*(OneNSE(Z,N,W,df) - OneNSE(Z,N+1,W,df))/2
    except:
        return None

def P3PointOED(Z,N,W,df):
    try:
        return (-1)**Z*(OnePSE(Z,N,W,df) - OnePSE(Z+1,N,W,df))/2
    except:
        return None

def SNESplitting(Z,N,W,df):
    try:
        return (-1)**N*(OneNSE(Z,N,W,df) - OneNSE(Z,N+2,W,df))
    except:
        return None

def SPESplitting(Z,N,W,df):
    try:
        return (-1)**Z*(OnePSE(Z,N,W,df) - OnePSE(Z+2,N,W,df))
    except:
        return None

def WignerEC(Z,N,W,df):
    if Z!=N:
        return None
    try:
        return DoubleMDiff(Z,N,W,df)-(DoubleMDiff(Z-2,N,W,df) + DoubleMDiff(Z,N+2,W,df))/2
    except:
        return None
    
def BEperA(Z,N,W,df):
    try:
        return BE(Z,N,W,df)/(Z+N)
    except:
        return None
```

```python
def uBE(Z,N,W,df):
    try:
        return float(df[(df['Z']==Z) & (df['N']==N) & (df['Wigner']==W)]['uBE'])
    except:
        return None

def uOneNSE(Z,N,W,df):
    try:
        return uBE(Z,N,W,df) + uBE(Z,N-1,W,df)
    except:
        return None
    
def uOnePSE(Z,N,W,df):
    try:
        return uBE(Z,N,W,df) + uBE(Z-1,N,W,df)
    except:
        return None

def uTwoNSE(Z,N,W,df):
    try:
        return uBE(Z,N,W,df) + uBE(Z,N-2,W,df)
    except:
        return None

def uTwoPSE(Z,N,W,df):
    try:
        return uBE(Z,N,W,df) + uBE(Z-2,N,W,df)
    except:
        return None

def uAlphaSE(Z,N,W,df):
    try:
        return uBE(Z,N,W,df) + uBE(Z-2,N-2,W,df)
    except:
        return None

def uTwoPSGap(Z,N,W,df):
    try:
        return uTwoPSE(Z,N,W,df) + uTwoPSE(Z+2,N,W,df)
    except:
        return None

def uTwoNSGap(Z,N,W,df):
    try:
        return uTwoNSE(Z,N,W,df) + uTwoNSE(Z,N+2,W,df)
    except:
        return None

def uDoubleMDiff(Z,N,W,df):
    try:
        return (uTwoPSE(Z,N,W,df) + uTwoPSE(Z,N-2,W,df))/4
    except:
        return None

def uN3PointOED(Z,N,W,df):
    try:
        return (uOneNSE(Z,N,W,df) + uOneNSE(Z,N+1,W,df))/2
    except:
        return None

def uP3PointOED(Z,N,W,df):
    try:
        return (uOnePSE(Z,N,W,df) + uOnePSE(Z+1,N,W,df))/2
    except:
        return None

def uSNESplitting(Z,N,W,df):
    try:
        return (uOneNSE(Z,N,W,df) + uOneNSE(Z,N+2,W,df))
    except:
        return None

def uSPESplitting(Z,N,W,df):
    try:
        return (uOnePSE(Z,N,W,df) + uOnePSE(Z+2,N,W,df))
    except:
        return None

def uWignerEC(Z,N,W,df):
    if Z!=N:
        return None
    try:
        return uDoubleMDiff(Z,N,W,df)+(uDoubleMDiff(Z-2,N,W,df) + uDoubleMDiff(Z,N+2,W,df))/2
    except:
        return None
    
def uBEperA(Z,N,W,df):
    try:
        return uBE(Z,N,W,df)/(Z+N)
    except:
        return None
```

```python
def eBE(Z,N,W,df):
    try:
        return float(df[(df['Z']==Z) & (df['N']==N) & (df['Wigner']==W)]['eBE'])
    except:
        return None

def eOneNSE(Z,N,W,df):
    try:
        return bool(eBE(Z,N,W,df) + eBE(Z,N-1,W,df))
    except:
        return None
    
def eOnePSE(Z,N,W,df):
    try:
        return bool(eBE(Z,N,W,df) + eBE(Z-1,N,W,df))
    except:
        return None

def eTwoNSE(Z,N,W,df):
    try:
        return bool(eBE(Z,N,W,df) + eBE(Z,N-2,W,df))
    except:
        return None

def eTwoPSE(Z,N,W,df):
    try:
        return bool(eBE(Z,N,W,df) + eBE(Z-2,N,W,df))
    except:
        return None

def eAlphaSE(Z,N,W,df):
    try:
        return bool(eBE(Z,N,W,df) + eBE(Z-2,N-2,W,df))
    except:
        return None

def eTwoPSGap(Z,N,W,df):
    try:
        return bool(eTwoPSE(Z,N,W,df) + eTwoPSE(Z+2,N,W,df))
    except:
        return None

def eTwoNSGap(Z,N,W,df):
    try:
        return bool(eTwoNSE(Z,N,W,df) + eTwoNSE(Z,N+2,W,df))
    except:
        return None

def eDoubleMDiff(Z,N,W,df):
    try:
        return bool(eTwoPSE(Z,N,W,df) + eTwoPSE(Z,N-2,W,df))
    except:
        return None

def eN3PointOED(Z,N,W,df):
    try:
        return bool(eOneNSE(Z,N,W,df) + eOneNSE(Z,N+1,W,df))
    except:
        return None

def eP3PointOED(Z,N,W,df):
    try:
        return bool(eOnePSE(Z,N,W,df) + eOnePSE(Z+1,N,W,df))
    except:
        return None

def eSNESplitting(Z,N,W,df):
    try:
        return bool(eOneNSE(Z,N,W,df) + eOneNSE(Z,N+2,W,df))
    except:
        return None

def eSPESplitting(Z,N,W,df):
    try:
        return bool(eOnePSE(Z,N,W,df) + eOnePSE(Z+2,N,W,df))
    except:
        return None

def eWignerEC(Z,N,W,df):
    if Z!=N:
        return None
    try:
        return bool(eDoubleMDiff(Z,N,W,df)+eDoubleMDiff(Z-2,N,W,df)+eDoubleMDiff(Z,N+2,W,df))
    except:
        return None
    
def eBEperA(Z,N,W,df):
    try:
        return bool(eBE(Z,N,W,df))
    except:
        return None
```

```python
def Wig1(Z, N):
    return (1.8*np.exp(-380*((N-Z)/(N+Z))**2))-(.84*abs(N-Z)*np.exp(-(((N+Z)/26)**2)))

def Wig2(Z, N):
    return -47*(abs(N-Z)/(N+Z))
```

```python
#Collect all raw exp data into each individual df

covar = [ME2, MEdelta, PC1, NL3S]

skm = [SKMS, SKP, SLY4, SV, UNEDF0, UNEDF1]
skm_even = [SKMS_even, SKP_even, SLY4_even, SV_even, UNEDF0_even, UNEDF1_even]
skm_merged = []
for i in range(len(skm)):
    skm_merged.append(skm[i].combine_first(skm_even[i]))

raw = [EXP]

for c in covar:
    raw.append(c)
for s in skm_merged:
    raw.append(s)
```

```python
filtered = []
for i, d in enumerate(raw):
    if i == 0:
        df = d.loc[:,['Z', 'N', 'Binding_Energy_(MeV)', 'Estimated', 'Uncertainty']]
    else:
        df = d.loc[:,['Z', 'N', 'Binding_Energy_(MeV)']]
        df['Binding_Energy_(MeV)'] = df['Binding_Energy_(MeV)'] * -1
    filtered.append(df)
```

```python
# convet units from KeV to MeV
filtered[0]['Binding_Energy_(MeV)'] = filtered[0]['Binding_Energy_(MeV)']/1000
filtered[0]['Uncertainty'] = filtered[0]['Uncertainty']/1000
```

```python
def Identity(Z, N):
    return 0
wigs = [Identity, Wig1, Wig2]
wigner_dfs =[]
for i, df in enumerate(filtered):
    if i == 0:
        new_df = pd.DataFrame(columns=['Z', 'N', 'Wigner', 'BE', 'eBE', 'uBE'])
        for wig in range(3):
            for rowi in range(len(df.index)):
                newBE = df.loc[rowi, 'Binding_Energy_(MeV)'] - wigs[wig](df.iloc[rowi,0], df.iloc[rowi,1])
                new_df.loc[len(new_df.index)] = [df.loc[rowi,'Z'], df.loc[rowi,'N'], wig, newBE, df.loc[rowi,'Estimated'], df.loc[rowi,'Uncertainty']]
    else:
        new_df = pd.DataFrame(columns=['Z', 'N', 'Wigner', 'BE'])
        for wig in range(3):
            for rowi in range(len(df.index)):
                newBE = df.loc[rowi, 'Binding_Energy_(MeV)'] - wigs[wig](df.iloc[rowi,0], df.iloc[rowi,1])
                new_df.loc[len(new_df.index)] = [df.loc[rowi,'Z'], df.loc[rowi,'N'], wig, newBE]
    wigner_dfs.append(new_df)
```

```python
mnames = ['EXP', 'ME2', 'MEdelta', 'PC1', 'NL3S', 'SKMS', 'SKP', 'SLY4', 'SV', 'UNEDF0', 'UNEDF1']
qfuncs = [OneNSE, OnePSE, TwoNSE, TwoPSE, AlphaSE, TwoNSGap, TwoPSGap, DoubleMDiff, N3PointOED, P3PointOED, 
          SNESplitting, SPESplitting, WignerEC, BEperA, uOneNSE, uOnePSE, uTwoNSE, uTwoPSE, uAlphaSE, uTwoNSGap, 
          uTwoPSGap, uDoubleMDiff, uN3PointOED, uP3PointOED, uSNESplitting, uSPESplitting, uWignerEC, uBEperA, 
          eOneNSE, eOnePSE, eTwoNSE, eTwoPSE, eAlphaSE, eTwoNSGap, eTwoPSGap, eDoubleMDiff, 
          eN3PointOED, eP3PointOED, eSNESplitting, eSPESplitting, eWignerEC, eBEperA]
qnames = ['OneNSE', 'OnePSE', 'TwoNSE', 'TwoPSE', 'AlphaSE', 'TwoNSGap', 'TwoPSGap', 'DoubleMDiff', 'N3PointOED', 
          'P3PointOED', 'SNESplitting', 'SPESplitting', 'WignerEC', 'BE/A', 'uOneNSE', 'uOnePSE', 'uTwoNSE', 
          'uTwoPSE', 'uAlphaSE', 'uTwoNSGap', 'uTwoPSGap', 'uDoubleMDiff', 'uN3PointOED', 
          'uP3PointOED', 'uSNESplitting', 'uSPESplitting', 'uWignerEC', 'uBE/A', 'eOneNSE', 
          'eOnePSE', 'eTwoNSE', 'eTwoPSE', 'eAlphaSE', 'eTwoNSGap', 'eTwoPSGap', 'eDoubleMDiff', 
          'eN3PointOED', 'eP3PointOED', 'eSNESplitting', 'eSPESplitting', 'eWignerEC', 'eBE/A']

quan_dfs = []
for m, model in enumerate(wigner_dfs):
    if m == 1:
        qfuncs, qnames = qfuncs[:14], qnames[:14]
    print(str(m) + ' checkpoint')
    df = pd.DataFrame(columns=qnames, index=range(len(model.index)))
    for rowi in range(len(model.index)):
        df.loc[rowi] = [q(model.loc[rowi, 'Z'], model.loc[rowi, 'N'], model.loc[rowi, 'Wigner'], model) for q in qfuncs]
    quan_dfs.append(df)
```

```python
final_dfs = [pd.concat([wigner_dfs[i], quan_dfs[i]], axis=1) for i in range(len(wigner_dfs))]
```

```python
for i in range(len(final_dfs)):
    final_dfs[i].to_hdf('6-8-23.h5', mnames[i])
```
